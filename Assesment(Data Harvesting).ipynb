I have used news api and "All articles mentioning Apple" Data 

```python
#Importing some usefull libraries
from newsapi import NewsApiClient
import pandas as pd
import json
import requests
```

Getting the URL from newsapi(After generating key) of all articles monitoring apple.

```python
URL='https://newsapi.org/v2/everything?q=apple&from=2021-07-07&to=2021-07-07&sortBy=popularity&apiKey=9a4801fc0b5647ffbc74e7285344f7b3'
data = requests.get(URL).json()
print(data)
type(data)
```

Here we have 2088 of data result , and type of the data is dictonary.

```python
#As data in dict. type we can check its key
print(data.keys())
```

```python
#we can check any  article 
print(data['articles'][1])
```

```python
#We can get key value for article 
for key , value in data['articles'][1].items():
    print(f"\n{key.ljust(15)} {value}")
```
